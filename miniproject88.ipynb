## Analyzing Tips Data


////Objective: Explore the relationships between tip amounts, total bill, and dining
details//

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

///Load data://

```python
tips=sns.load_dataset("tips")
print(tips.head())
```

///Relational Plot://

```python
plt.figure(figsize=(6,6))
sns.relplot(x="total_bill",y="tip",hue="day",style="time",data=tips)
plt.title("relationship between total bill and tips")
plt.show()
```

///Categorical Analysis://

```python
plt.figure(figsize=(6,6))
sns.violinplot(x="day",y="tip",data=tips)
plt.title("Tip distribution by day")
plt.show()
```

///HeatMap for correlation://

```python
num=tips.select_dtypes(include=["float64","int64"])
corr_matrix=num.corr()
plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap - Tips Dataset")
plt.show()
```
