## Student Test Score Analysis

///Objective: Explore a student test score dataset to find relationships
between scores and demographic factors.//

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

///Load Data://

```python
data={'gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'study_group': ['A', 'A', 'B', 'B', 'C', 'C', 'A', 'B'],
    'math_score': [78, 85, 69, 92, 73, 88, 81, 79],
    'science_score': [82, 90, 72, 95, 75, 91, 84, 80]}
df=pd.DataFrame(data)
```

///Box plot to compare scores by gender

```python
df["total_score"]=df["math_score"]+df["science_score"]
print(df)
score=df.groupby('gender')["total_score"].sum().reset_index()
print(score)
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x="total_score",y="gender",data=df)
plt.title("scores by gender")
plt.show()
```

///violinplot to see score for study groups//

```python
plt.figure(figsize=(6,6))
sns.violinplot(x="study_group",y="total_score",data=df)
plt.title("Score for different study groups")
plt.show()
```

// Scatterplot  Relationship Between Two Subjects//

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='math_score', y='science_score', hue='gender', data=df)
plt.title('Relationship Between Math and Science Scores')
plt.xlabel('Math Score')
plt.ylabel('Science Score')
plt.show()
```
