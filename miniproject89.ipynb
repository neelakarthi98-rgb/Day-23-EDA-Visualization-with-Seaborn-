## Sales Performance Visualization

/// Use a sales dataset to visualize sales performance//

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

///Load data://

```python
data = {
    'date': ['2024-01-10', '2024-02-10', '2024-03-10', '2024-01-08', '2024-02-15', '2024-03-22'],
    'region': ['East', 'East', 'East', 'West', 'West', 'West'],
    'category': ['Electronics', 'Electronics', 'Electronics', 'Furniture', 'Furniture', 'Furniture'],
    'sales': [3000, 4500, 5000, 2800, 4000, 4700]
}
df=pd.DataFrame(data)
```

```python
### convert date column to datetime
df["date"]=pd.to_datetime(df["date"])
##extract month name
df['month']=df['date'].dt.strftime('%b')
print(df)
```

//Line plot to show sales trend over time//

```python
plt.figure(figsize=(8, 5))
sns.lineplot(data=df, x='date', y='sales', marker='o')
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

///Bar plot to show sales across different region//

```python
plt.figure(figsize=(7, 5))
sns.barplot(data=df, x='region', y='sales', estimator='sum')
plt.title("Total Sales by Region")
plt.ylabel("Total Sales")
plt.show()
```

////Heat map sales by month and category

```python

pivot = df.pivot_table(
    index='month',
    columns='category',
    values='sales',
    aggfunc='sum'
)

plt.figure(figsize=(6, 5))
sns.heatmap(pivot, annot=True, cmap='Blues')
plt.title("Sales by Month and Category (Heatmap)")
plt.show()
```
