## Titanic Survival Analysis

////Objective: Use visualization to explore the factors influencing passenger survival
on the Titanic///

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

///Load data://

```python
titanic=sns.load_dataset("titanic")
head=titanic.head()
print(head)
```

///Univariate Analysis://

```python
plt.Figure(figsize=(10,6))
sns.histplot(data=titanic,x="age",kde=True)
plt.title("Distribution of passenger Age")
plt.show()
```

///Bivariate and Categorical Analysis://


```python
###Survival by class
plt.Figure(figsize=(6,6))
sns.barplot(x='class',y='survived',data=titanic)
plt.title("Survival Rate by passenger class")
plt.show()
```

```python
###survival by gender
plt.figure(figsize=(5,5))
sns.countplot(x='sex',hue="survived",data=titanic)
plt.title("survival count by gender")
plt.show()
```

////Multivariate Analysis://

```python
plt.Figure(figsize=(10,6))
sns.scatterplot(x="age",y="fare",hue="survived",data=titanic)
plt.title("Fare vs Age colured by survived")
plt.show()
```
